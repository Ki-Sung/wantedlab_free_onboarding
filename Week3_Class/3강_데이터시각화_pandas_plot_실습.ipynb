# Pandas의 DataFrame의 .plot으로 그래프 그리기

## Reference : https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html

* ‘bar’ or ‘barh’ for bar plots
* ‘hist’ for histogram
* ‘box’ for boxplot
* ‘kde’ or ‘density’ for density plots
* ‘area’ for area plots
* ‘scatter’ for scatter plots
* ‘hexbin’ for hexagonal bin plots
* ‘pie’ for pie plots

```python
import pandas as pd
import numpy as np
```

### Line Plot 그리기

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))
ts = ts.cumsum()  # 누적합 cumsum () 사용
ts.plot(figsize=(12,8));
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list("ABCD"))
df = df.cumsum()
df.plot();
```

```python
df
```

### Bar Plot 그리기

```python
df.iloc[5]
```

```python
df.iloc[5].plot(kind="bar");
```

```python
df.iloc[5].plot(kind="barh");
```

### Hist plot 그리기

```python
df4 = pd.DataFrame(
    {
        "a": np.random.randn(1000) + 1,
        "b": np.random.randn(1000),
        "c": np.random.randn(1000) - 1,
     },
     columns=["a", "b", "c"],
     )
df4.plot(kind="hist", alpha=0.5)
```

```python
df4
```

### Box Plot 그리기

```python
df = pd.DataFrame(np.random.rand(10, 5), columns=["A", "B", "C", "D", "E"])
df.plot(kind="box");
```

```python
df
```

### Area plot 그리기

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=["a", "b", "c", "d"])
df.plot(kind="area");
```

```python
df
```

### Scatter plot 그리기

```python
df = pd.DataFrame(np.random.rand(50, 4), columns=["a", "b", "c", "d"])
df.plot(kind="scatter", x="a", y="b");
```

```python
df
```

```python
df.plot(kind="scatter", x="a", y="c");
```

### Pie plot 그리기

```python
series = pd.Series(3 * np.random.rand(4), index=["a", "b", "c", "d"], name="series")
series.plot(kind="pie",figsize=(6, 6))
```

```python
series
```

```python
df = pd.DataFrame(
    3 * np.random.rand(4, 2), index=["a", "b", "c", "d"], columns=["x", "y"]
    )
df.plot(kind="pie", subplots=True, figsize=(8, 4));
```
